```python
################### 4/5

# range 로 데이터 추출: 지정한 구간에서 정수의 리스트를 반환
# 좀 더 정확하게 보면 지정한 범위의 정수리스트를 반환하는 것이 아니라 제네리어터를 반환하는것
# iloc 속성은 제네레이터로 데이처를 추출할 수 없음 -> 제네레이터는 간단하게 리스트로 변환도 가능

small_range = list(range(5))
print(small_range)
print(type(small_range))
```

```python
subset = df.iloc[:,small_range]
print(subset.head())
```

```python
small_range = list(range(0,6,2))
subset = df.iloc[:,small_range]
print(subset.head())
```

```python
subset2 = df.iloc[:,3]
print(subset2.head())
```

```python
subset = df.iloc[[0,99,999],[0,3,5]]
print(subset.head())

# iloc 속성의 열 지정값으로 정수 리스트를 전달하는게 편해 보이지만 이렇게
# 코드를 작성하면 나중에 어떤 데이터를 추출하려고 했던것인지 파악이
# 어려울수도 있다. 그래서 보통 loc속성을 사용하여 열 지정값으로 열 이름을 전달하게 됨
```

```python
# lifeExp 열을 연도별로 그룹화하고 평균을 계산

# 데이터를 year로 그룹화
# lifeExp열의 평균을 계산

group_df = df.groupby('year')
print(type(group_df))
print(group_df)

group_year_lifeExp = group_df['lifeExp']
print(group_year_lifeExp)

# 연도별로 그룹화한 lifeExp
df.groupby(['year','continent'])[['lifeExp','gdpPercap']].mean()


```

```python
df.groupby(['year','continent'])[['lifeExp','gdpPercap']].mean()
```

```python
# 데이터를 시각화하면 데이터를 이해하거나 추이를 파악하거나 어떤 통찰력을 얻는데
# 아주 많은 도움이 된다. 

import matplotlib.pyplot as plt

group_year_lifeExp.mean().plot()

```
