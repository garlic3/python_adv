```python
# 데이터 프레임에 시리즈
new_row = pd.Series(['n1','n2','n3','n4'])
```

```python
pd.concat([df1, new_row])

# 시리즈가 새로운 행으로 추가 될것을 기대했지만 새로운 열로 삽입이 되면서
# 구조를 오히려 복잡하게 만들고 있음. NaN 많이 보임
# 행이 1개라도 반드시 데이터프레임에 담아서 연결
# 시리즈에는 열이름이 없음. 따라서 새로운 열로 간주하여 0이라는 이름의 열로 추가하게 된것

```

```python
new_row_df = pd.DataFrame([['n1','n2','n3','n4']], columns=['A','B','C','D'])
print(new_row_df)
```

```python
# 방법 1, 데이터 프레임으로 만들어서 데이터프레임 끼리 연결
pd.concat([df1, new_row_df])
```

```python
# 방법 2, 데이터 프레임으로 만들어서 데이터프레임 끼리 append 연결

df1.append(new_row_df, ignore_index=True)
```

```python
# 방법 3, 
data_dict = {'A':'n1', 'B':'n2', 'C':'n3', 'D':'n4'}

print(df1.append(data_dict, ignore_index=True))

# 마지막 행의 인덱스가 0 이 아닌 4로 변경
```

```python
# 열방향으로 데이터 연결

col_coucat = pd.concat([df1, df2, df3], axis=1)
col_coucat
```

```python
col_coucat['F'] = ['n1','n2','n3','n4']
col_coucat
```

```python
# 열 이름의 일부가 서로 다른 데이터 프레임을 연결하면 
df2.columns = ['E','F','G','H']
df3.columns = ['A','C','F','H']
```

```python
row_concat2 = pd.concat([df1, df2, df3])
row_concat2

# 새롭게 열이름을 부여하고 3개의 데이터 프레임을 concat으로 연결
# 열이름이 정렬되어 연결되었고 데이터프레임에 없는 열이름의 데이터는 전부 누락값으로 처리됨
# 데이터프레임의 공통열만 골라서 연결하면 누락값이 발생하지 않음
# 현재 df1, df2, df3 모두의 공통열은 없음

```

```python
# inner join 내부 조인 - 둘 이상의 df 에서 조건에 맞는 행을 연결 
# outer join 외부 조인 - 왼쪽 외부조인 left outer join(왼쪽 df를 모두), 오른쪽 외부조인 right outer join(오른쪽 df를 모두), 완전 외부 조인 full outer join(양쪽 df를 모두)
```

```python
# 판다스는 데이터 연결 전용 메서드인 merge를 제공

person = pd.read_csv('survey_person.csv')
site = pd.read_csv('survey_site.csv')
survey = pd.read_csv('survey_survey.csv')
visited = pd.read_csv('survey_visited.csv')
```

```python
merge1 = site.merge(visited_subset, left_on='name', right_on='site')
merge1
```

```python
merge2 = site.merge(visited, left_on='name', right_on='site')
merge2
```

```python
merge3 = person.merge(survey, left_on='ident', right_on='person')
merge3
```

```python
# 누락값 - 결측치 , 대부분의 데이터에는 누락값이 존재 
# NaN(NAN, nan)

from numpy import NaN, NAN, nan
```

```python
print(NaN == True)
print(NaN == False)
print(NaN == 0)
print(NaN == '')
print(NaN == NaN)
```

```python
# 확인하는 매서드 isnull 을 이용

print(pd.isnull(NaN))
print(pd.isnull(NAN))
print(pd.isnull(nan))
```

```python
print(pd.notnull(NaN))
print(pd.notnull(NAN))
print(pd.notnull(nan))
```
