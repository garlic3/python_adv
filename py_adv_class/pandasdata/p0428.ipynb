```python
ebola.fillna(method='ffill')

# 결측치가 나타나기 전의 값으로 변경함
# 예를 들면 6행의 결측치가 있을 때 5행의 값을 사용하여 결측치를 처리
# 0, 1행의 경우에는 전의 값이 없기 때문에 결측치가 그대로 남아있게 됨
```

```python
ebola2 = pd.read_csv('country_timeseries.csv')
ebola2.fillna(method='ffill').iloc[0:10, 0:5]
```

```python
# bfill 로 지정하면 결측치가 나타난 이후의 첫번째 값으로 앞쪽의 결측치가 모두 변경됨
# 즉 ffill의 반대 방향으로 처리. 하지만 이 방법도 마지막 값이 누락값인 경우에는 처리하지 못함 
ebola3 = pd.read_csv('country_timeseries.csv')
ebola3.fillna(method='bfill').iloc[0:10, 0:5]
```

```python
ebola.interpolate().iloc[0:10, 0:5]
# 위에섯의 두가지 방법 ffill bfill는 데이터 프레임의 값을 그대로 사용하여 결측치를 처리했으나 interpolate 매서드는 누락값 양쪽에 있는 값을 이용하여
# 중간값을 구해서 결측치를 처리함 이렇게 하면 데이터 프레임이 일정한 간격을 유지하고 있는 것처럼 수정할 수 있음
```
